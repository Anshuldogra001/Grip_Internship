```python
# Import all necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets
from plotnine import *
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
s_data = pd.read_csv("SampleSuperstore.csv")
```

```python
s_data.isnull().sum()
```

```python
# Chaecking the dupilication in data
s_data.duplicated().sum()
```

```python
# Removing the duplicates
s_data.drop_duplicates()
```

```python
s_data.describe()
```

```python
#Deleting the Variable.
col=['Postal Code']
s_data=s_data.drop(columns=col,axis=1)
```

```python
#Correlation Between Variables.
s_data.corr()
```

```python
# Covariance of columns
s_data.cov()
```

```python
s_data.head()
```

#Exploratory Data Analysis


```python
s_data.plot(x='Sales', y='Profit', style='o')  
plt.title('Sales vs Profit')  
plt.xlabel('Sales')  
plt.ylabel('Profit')  
plt.show()
```

```python
#Loss coming after providing discount
sns.stripplot(x='Discount', y='Profit', data=s_data)
plt.title('Graph')
 
# function to show plot
plt.show()
```

```python
s_data.hist(bins=50 ,figsize=(16,13))
plt.show();
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Category', data=s_data)
plt.show()
```

```python
plt.figure(figsize=(17,6))
sns.countplot(x='Sub-Category', data=s_data)
plt.show()
```

```python
df=s_data.groupby(['Segment','Category']).size()
df=df.unstack()
df.plot(kind='bar')
```

```python
df=s_data.groupby(['Segment','Sub-Category']).size()
df=df.unstack()
df.plot(kind='bar', figsize=(15, 10))
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.countplot(x="Region", data=s_data)
 
# giving title to the plot
plt.title('Graph')
 
# function to show plot
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Segment', data=s_data)
plt.show()
```

```python
df=s_data.groupby(['Segment','Region']).size()
df=df.unstack()
df.plot(kind='bar')
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=s_data['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python

Profit_plot = (ggplot(s_data, aes(x='Sub-Category', y='Profit', fill='Sub-Category')) + geom_col() + coord_flip()
+ scale_fill_brewer(type='div', palette="Spectral") + theme_classic() + ggtitle('Pie Chart'))

display(Profit_plot)
```

```python
df=s_data.groupby(['Ship Mode','Category']).size()
df=df.unstack()
df.plot(kind='bar', figsize=(13,6))
```

```python
df=s_data.groupby(['Ship Mode','Sub-Category']).size()
df=df.unstack()
df.plot(kind='bar', figsize=(15,9))
```

##plotting a graph for Quantity VS Sales data

```python
sns.set(style="whitegrid")
plt.figure(2, figsize=(20,15))
sns.barplot(x='Quantity',y='Sales', data=s_data, palette='Spectral')
plt.suptitle('Pie Consumption Patterns in the United States', fontsize=16)
plt.show()
```
